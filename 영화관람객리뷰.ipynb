```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
for j in range(1,10) :
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=187320&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(j))
    soup = bs(res.text, "lxml")
    
    viewer = soup.select('span.ico_viewer')
    for i in viewer :
        i.extract()
        
    spo = soup.select("span.text_spo._text_spo")
    for i in spo :
        i.extract()
        
    review = soup.select("div.score_reple > p")
    for i in review :
        print(i.text.strip())
```

```python
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=187320&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
# 데이터 접근이 불가능한 이유는
# 화면에는 보이지만 실제 데이터는 다른 주소에 저장 되어있다.
# <iframe>태그를 통해 정보를 보여주고만 있다.
# 실제 데이터가 있는 주소로 찾아가기
review = soup.select("div.score_reple > p")
#1 불필요한 개행문자(이스케이프코드) 삭제하는 방법 strip()
for i in review :
    print(i.text.strip())
```

```python
#2 불필요한 요소를 추출해서 제거하는 방법 extract()
# 사용시 원본 데이터에 영향을 줌
# 불필요한 관람객 글자를 삭제시키기 위해서
# 관람객 '요소'만 수집
viewer = soup.select('span.ico_viewer')

for i in viewer :
    i.extract()
```
